```python
#importing necessary packages
import csv
from pathlib import Path
```

```python
#setting a variable to store the file path 
csvpath = Path("../../Resources/budget_data.csv")

#opening up the file and reading from it
with open(csvpath, "r") as csvfile:
    csvreader = csv.DictReader(csvfile, delimiter = ",")
    
    #declaring necessary variables to output
    total_months = 0
    total_profits = 0
    average_change = 0
    max_increase = None
    max_increase_date = None
    max_decrease = None
    max_decrease_date = None
    
    #declaring calculation variables
    prev_value = None
    total_change = 0 
    
    #for loop to process data
    for row in csvreader:
        # for more convenient processing
        current_value = int(row["Profit/Losses"]) 
        current_date = row["Date"]
        
        total_months +=1 #each line = new month
        total_profits += current_value #cumulates total profits
        
        
        if prev_value == None: #establishing base case where there is no previous value
            prev_value = current_value
        else: 
            #setting up all variables to be used in the loop
            current_change = current_value - prev_value 
            prev_value = current_value
            total_change += current_change
            
            #for min/max, set base case and increment case
            if max_increase == None: #base case when no previous max exists
                max_increase = current_value 
                max_increase_date = current_date
            else: #if max is already set, then checks if new max
                if current_change > max_increase:
                    max_increase = current_change
                    max_increase_date = current_date
            
            if max_decrease == None: #base case where no previous min exists
                max_decrease = current_value
                max_decrease_date = current_date
            else: #if min is already set, then checks if new min 
                if current_change < max_decrease:
                    max_decrease = current_change
                    max_decrease_date = current_date
       
    
    file1 = open("output_file.txt","w") #opens a new output file
    #then writes the entire line 
    file1.write(f"Financial Analysis \n ---------------------------- \nTotal Months: {total_months} \nTotal: ${total_profits} \nAverage Change: ${total_change/(total_months-1):.2f} \nGreatest Increase in Profits: {max_increase_date} (${max_increase})\nGreatest Decrease in Profits: {max_decrease_date} (${max_decrease})")   

    
    
    # sample output
    '''
    Financial Analysis
----------------------------
Total Months: 86
Total: $38382578
Average  Change: $-2315.12
Greatest Increase in Profits: Feb-2012 ($1926159)
Greatest Decrease in Profits: Sep-2013 ($-2196167)

    '''







    
```
